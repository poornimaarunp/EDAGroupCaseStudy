```python
# summing up the missing values (column-wise)
round(100*(df_loan.isnull().sum()/len(df_loan.index)), 2)
```

```python
# removing columns with all missing values
df_loan_new=df_loan.dropna(axis=1, how='all')
```

```python
df_loan_new.shape
```

```python
# summing up the missing values (column-wise)
round(100*(df_loan_new.isnull().sum()/len(df_loan_new.index)), 2)
```

```python
# deleting columns with more than 30% missing

df_loan_new = df_loan_new.drop('desc', axis=1)
df_loan_new = df_loan_new.drop('mths_since_last_delinq', axis=1)
df_loan_new = df_loan_new.drop('mths_since_last_record', axis=1)
df_loan_new = df_loan_new.drop('next_pymnt_d', axis=1)
```

```python
round(100*(df_loan_new.isnull().sum()/len(df_loan_new.index)), 2)
```

```python
# treating missing values in rows where >5

df_loan_new[df_loan_new.isnull().sum(axis=1) > 5]
```

```python
# Data Cleaning
# Cleaning employee Length

df_loan_new['emp_length_clean'] = df_loan_new.emp_length.str.replace('+','')
df_loan_new['emp_length_clean'] = df_loan_new.emp_length_clean.str.replace('< 1','0')
df_loan_new['emp_length_clean'] = df_loan_new.emp_length_clean.str.replace('years','')
df_loan_new['emp_length_clean'] = df_loan_new.emp_length_clean.str.replace('year','')

```

```python
df_loan_new.emp_length_clean.unique()
```

```python
df_loan_new['emp_length_clean'] = df_loan_new.emp_length_clean.map(float)
```

```python
# removing months from term

df_loan_new['term']=df_loan_new['term'].map(lambda x: x.strip('months'))
df_loan_new.head(30)
```
